```python
img_width = 150
img_height = 150

train_data_dir = 'C:\\Users\\YOGA\\Downloads\\Dataset\\Dataset\\Train'
validation_data_dir = 'C:\\Users\\YOGA\\Downloads\\Dataset\\Dataset\\Val'
train_samples = 120
validation_samples = 30
epochs = 5
batch_size = 20

# Check for TensorFlow or Thieno
if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation="relu",input_shape=(256,256,3)))

model.add(Conv2D(64,(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(64,(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(128,(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(128,(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(1,activation="sigmoid"))

model.compile(loss=keras.losses.binary_crossentropy,optimizer = "adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
train_datagen = image.ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
)
test_dataset = image.ImageDataGenerator(rescale = 1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    'C:\\Users\\YOGA\\Downloads\\Dataset\\Dataset\\Train',
    target_size = (256,256),
    batch_size = 32,
    class_mode = 'binary'
)
```

```python
Classes=train_generator.class_indices
Classes
```

```python
validation_generator = test_dataset.flow_from_directory(
    'C:\\Users\\YOGA\\Downloads\\Dataset\\Dataset\\val',
    target_size = (256,256),
    batch_size = 32,
    class_mode = 'binary'
)
```

```python
hist = model.fit_generator(
    train_generator,
    steps_per_epoch = 8,
    epochs = 10,
    validation_data = validation_generator,
    validation_steps = 2
)
```

```python
print(hist.history.keys())
# summarize history for accuracy
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
def preprocess(img_path):
    img1 = image.load_img(img_path, target_size=(256, 256,3))
    img = image.img_to_array(img1)
    img = img/255
    return np.expand_dims(img, axis=0)
   
```
